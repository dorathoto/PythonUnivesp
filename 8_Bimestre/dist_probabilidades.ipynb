```python
import matplotlib.pyplot as plt
```

```python
def fact(n):
    fat = 1
    i = 2
    while i <= n:
        fat = fat*i
        i = i + 1

    return fat
```

```python
px = []

for k in range(6):
    n = 5
    p = (fact(n) / (fact(k) * fact(n-k))) * 0.2**k * 0.8**(n-k)
    px.append(p)
    print(f'x = {k} -> {p:.5}')    
```

```python
plt.bar(range(6), px)
plt.show()
```
